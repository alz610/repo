```python
import numpy as np
from scipy.linalg import solve_banded
```

```python
%%time
n = 100000000
a_, c_, b_, f_ = 1, 3, 1, 1

ab = np.zeros((3, n))

ab[((0, 2), (0, n - 1))] = 0

ab[0] = b_
ab[1] = c_
ab[2] = a_

f = np.full(n, f_)

x = solve_banded((1, 1), ab, f)
```
